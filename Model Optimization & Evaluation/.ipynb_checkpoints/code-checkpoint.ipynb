```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("titanic")
```

```python
df.drop(["deck", "embark_town", "alive", "class", "who", "adult_male"], axis=1, inplace=True)
```

```python
df["age"].fillna(df["age"].mean(), inplace=True)
```

```python
df.dropna(subset=["embarked"], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['sex'] = le.fit_transform(df['sex'])
df["embarked"] = le.fit_transform(df["embarked"])
```

```python
df = df.astype(int)
```

```python
X = df.drop("survived", axis=1)
y = df["survived"]
```

# Cross Validation

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_results = cross_val_score(model, X,y, cv = 5, scoring='accuracy')
```

```python
print(cross_val_results)
```

# Ensemble Learning

# 1. Stacking

```python

from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df3 = sns.load_dataset('iris')
```

```python
df3.head()
```

```python
X3 = df3.drop('species', axis=1)
y3 = df3['species']
```

```python

le = LabelEncoder()
y_encoded = le.fit_transform(y3)
     
```

```python
X_train, X_test, y_train, y_test = train_test_split(X3, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python

base_learners = [
    ('dt', DecisionTreeClassifier(random_state=42)),
    ('svc', SVC(probability=True, kernel='rbf', random_state=42)),
    ('lr', LogisticRegression(max_iter=1000))
]
```

```python
meta_learner = LogisticRegression(max_iter=1000)
```

```python
stacking_clf = StackingClassifier(
    estimators=base_learners,
    final_estimator=meta_learner,
    cv=5
)
```

```python
stacking_clf.fit(X_train, y_train)
```

```python
y_pred = stacking_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
accuracy
```
