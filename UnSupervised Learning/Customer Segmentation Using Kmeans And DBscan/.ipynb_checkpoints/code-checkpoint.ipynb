```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import RobustScaler
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA

sns.set(style="whitegrid")

# Load dataset
df = pd.read_csv("dataset.csv")

# Basic data cleaning
df = df.dropna(subset=["CustomerID"])
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# Remove cancelled or invalid transactions
df = df[(df["Quantity"] > 0) & (df["UnitPrice"] > 0)]

# Create total transaction value
df["TotalPrice"] = df["Quantity"] * df["UnitPrice"]

# -----------------------
# Feature engineering (RFM)
# -----------------------

reference_date = df["InvoiceDate"].max()

rfm = df.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (reference_date - x.max()).days,  # Recency
    "InvoiceNo": "nunique",                                    # Frequency
    "TotalPrice": "sum"                                       # Monetary
})

rfm.columns = ["Recency", "Frequency", "Monetary"]

# Log transform to reduce skewness
rfm_log = np.log1p(rfm)

# Scale features (robust to outliers)
scaler = RobustScaler()
rfm_scaled = scaler.fit_transform(rfm_log)

# -----------------------
# Model selection (KMeans)
# -----------------------

inertia = []
sil_scores = []
K = range(2, 11)

for k in K:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(rfm_scaled)
    inertia.append(km.inertia_)
    sil_scores.append(silhouette_score(rfm_scaled, labels))

# Elbow method
plt.figure(figsize=(6,4))
plt.plot(K, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()

# Silhouette analysis
plt.figure(figsize=(6,4))
plt.plot(K, sil_scores, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Analysis")
plt.show()

# -----------------------
# Final KMeans clustering
# -----------------------

kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
rfm["KMeans_Cluster"] = kmeans.fit_predict(rfm_scaled)

# -----------------------
# PCA for visualization
# -----------------------

pca = PCA(n_components=2, random_state=42)
rfm_pca = pca.fit_transform(rfm_scaled)

rfm["PCA1"] = rfm_pca[:, 0]
rfm["PCA2"] = rfm_pca[:, 1]

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Explained:", pca.explained_variance_ratio_.sum())

# KMeans clusters in PCA space
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="PCA1",
    y="PCA2",
    hue="KMeans_Cluster",
    data=rfm,
    palette="tab10",
    s=60
)

plt.axhline(0, color="black", linestyle="--", linewidth=1)
plt.axvline(0, color="black", linestyle="--", linewidth=1)

plt.title("Customer Segmentation using KMeans (PCA View)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# -----------------------
# DBSCAN for outlier detection
# -----------------------

dbscan = DBSCAN(eps=0.5, min_samples=6)
rfm["DBSCAN_Cluster"] = dbscan.fit_predict(rfm_scaled)

print("\nDBSCAN Cluster Distribution:")
print(rfm["DBSCAN_Cluster"].value_counts())

# DBSCAN result in PCA space
plt.figure(figsize=(8,6))
sns.scatterplot(
    x="PCA1",
    y="PCA2",
    hue="DBSCAN_Cluster",
    data=rfm,
    palette="tab10",
    s=60
)

plt.axhline(0, color="black", linestyle="--", linewidth=1)
plt.axvline(0, color="black", linestyle="--", linewidth=1)

plt.title("Customer Segmentation using DBSCAN (PCA View)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

# -----------------------
# Cluster interpretation
# -----------------------

print("\nKMeans Cluster Summary (Mean RFM Values):")
print(
    rfm.groupby("KMeans_Cluster")[["Recency", "Frequency", "Monetary"]]
    .mean()
    .round(2)
)
```
